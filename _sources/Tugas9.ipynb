```python
dataset_url = ('/content/drive/MyDrive/Datamining/tugas/data_csv/messidor_features.arff')
```

```python
data = arff.loadarff(dataset_url)
df = pd.DataFrame(data[0])
df.head()
```

```python
df.shape
```

```python
col_names = []
for i in range(20):
    if i == 0:
        col_names.append('quality')
    if i == 1:
        col_names.append('prescreen')
    if i >= 2 and i <= 7:
        col_names.append('ma' + str(i))
    if i >= 8 and i <= 15:
        col_names.append('exudate' + str(i))
    if i == 16:
        col_names.append('euDist')
    if i == 17:
        col_names.append('diameter')
    if i == 18:
        col_names.append('amfm_class')
    if i == 19:
        col_names.append('label')
```

```python
df.columns = [col_names]
```

```python
df.head()
```

```python

X = df.drop(columns=["label"])
X.head()
```

```python
y=df['label'].values
y
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
```

```python
 y=le.transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

training, test = train_test_split(df, train_size = 0.8, test_size = 0.2, shuffle=False)
training_label, test_label = train_test_split(y, train_size = 0.8, test_size = 0.2, shuffle=False)

clf2 = GaussianNB()
clf2.fit(training, training_label)

probas = clf2.predict_proba(test)
```

```python
# percent_amount_of_test_data = / HUNDRED_PERCENT
percent_amount_of_test_data = 0.3
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = percent_amount_of_test_data, random_state=0)
```

```python
# classifier
gaussian = GaussianNB()
# 
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for Naive Bayes\n',cm)
print('accuracy_Naive Bayes: %.3f' %accuracy)
print('precision_Naive Bayes: %.3f' %precision)
print('recall_Naive Bayes: %.3f' %recall)
print('f1-score_random_Forest : %.3f' %f1)
```
