```python
# league code
code_league_for_cherbourg = "NL"
code_league_for_southampton = "AL"

# binary value
value_of_one = 1
value_of_zero = 0

def change_code_league_to_biner(leagueed):
    return value_of_one if leagueed == code_league_for_southampton else value_of_zero
```

```python
# Update all values of 'leagueed' series
df["leagueed"] = df["leagueed"].apply(change_code_league_to_biner)
```
